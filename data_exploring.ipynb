### Data visualization

```python
#Check the imbalance of the Class column
count = df.Class.value_counts()
count.plot.bar()
plt.xlabel('Class')
plt.ylabel('transaction times')
plt.title('Fraud vs Normal')
plt.text(0, count[0], count[0], fontsize=14, horizontalalignment='center', verticalalignment='top')
plt.text(1, count[1], count[1], fontsize=14, horizontalalignment='center', verticalalignment='bottom')
```

```python
#show how time compares across fraudulent and normal transactions. 
plt.style.use('seaborn')
plt.figure(figsize=(20,10))
fig, (ax1, ax2) = plt.subplots(2, 1)

ax1.hist(df.Time[df.Class == 1], bins = 35,color = 'y')
ax1.set_title('Fraud', fontsize =15  )
ax2.hist(df.Time[df.Class == 0], bins = 35,color = 'g')
ax2.set_title('Normal', fontsize =15 )
plt.tight_layout()
plt.xlabel('Time (in Seconds)',fontsize =12)
plt.ylabel('Number of Transactions',fontsize =12)
plt.show()

```

The "time" feature looks very similar for both types of transactions. Normal transactions show regularity and have some certain periodic distribution.fraudulent transactions are more evenly. 


```python
#show how Amount compares across fraudulent and normal transactions. 
plt.style.use('seaborn')
plt.figure(figsize=(20,10))
fig, (ax1, ax2) = plt.subplots(2, 1)

ax1.hist(df.Amount[df.Class == 1], bins = 35,color = 'y')
ax1.set_title('Fraud', fontsize =15  )
ax2.hist(df.Amount[df.Class == 0], bins = 35,color = 'g')
ax2.set_title('Normal', fontsize =15 )
plt.tight_layout()
plt.xlabel('Amount ($)',fontsize =12)
plt.ylabel('Number of Transactions',fontsize =12)
plt.yscale('log')
plt.show()
```

Most fraud transactions are less than $500. Most normal transactions are less than $10000. 

```python
#show how Amount compares with Time for fraudulent and normal transactions. 
plt.style.use('seaborn')
plt.figure(figsize=(20,10))
fig, (ax1, ax2) = plt.subplots(2, 1)

ax1.scatter(df.Time[df.Class == 1],df.Amount[df.Class == 1], color = 'y')
ax1.set_title('Fraud', fontsize =15  )
ax2.scatter(df.Time[df.Class == 0],df.Amount[df.Class == 0], color = 'g')
ax2.set_title('Normal', fontsize =15 )
plt.tight_layout()
plt.xlabel('Time (in Seconds)',fontsize =12)
plt.ylabel('Amount ($)',fontsize =12)
plt.show()
```

```python
#show V1-V28 and Amount, Class, Time features distribution
plt.style.use('seaborn')
plt.figure(figsize=(20,20))
for i in range(0,df.shape[1]-1):
    plt.subplot(8, 4, i+1)
    sns.distplot(df.iloc[:,i], kde=True,color = 'g')
    plt.tight_layout()
```
